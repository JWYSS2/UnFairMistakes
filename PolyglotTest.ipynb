```python
def lang_detect(text):
  try:
    temp = Detector(text)
    string = str(temp).split('\n')
    lang_1 = string[1].split()
    return {'name': lang_1[3], 'code': lang_1[5], 'confidence': lang_1[7], 'read bytes':  lang_1[10]}
  except:
    return {'name': 'unknown', 'code': '?', 'confidence': 0, 'read bytes':  None}
```
